<h4><b>OBSERVAÇÃO 01</b></h4><br>
Agora que obtemos os elementos contidos no objeto 'Select', devemos CONCATENAR ao endereço padrão das páginas de cada estado, substituindo a sigla para cada sigla.<br><br>
Observe o endereço do estado do AC:<br>
http://premiacao.obmep.org.br/16aobmep/verMenuAlunosPremiados-<b>AC</b>.htm
<br><br>
O que precisamos é criar uma função que realize um 'loop' e a cada ciclo substituir a sigla para criar a url de cada estado, para extração dos dados contidos lá.
<br><br><br><br>


---



```python
# Cria duas listas vazias
lista_de_urls = []
lista_de_siglas = []

# Cria loop para cada elemento dentro do objeto 'select'
for s1 in select:

  # Adquire o conteúdo de cada elemento em 'select -> s1'
  for s2 in s1:
    # Comando para testar um bloco de comandos e direcionar para o caso de erro
    try:

      # Obtém o conteúdo (caso exista) de cada elemento e concatena à url para cada estado.
      sigla = s2.get_text()

      # remove quebra de linha da string
      sigla = sigla.replace("\n", "")

      # Verifica se o conteúdo da string tem mais que ZERO caracteres
      if len(sigla) > 0:
        # Adiciona sigla à lista de siglas
        lista_de_siglas.append(sigla)

        # Adiciona a sigla do estado à string 'url_temp'
        url_temp = 'http://premiacao.obmep.org.br/16aobmep/verMenuAlunosPremiados-{}.htm'.format(sigla)

        print(url_temp)

        # Adiciona url_temp à lista
        lista_de_urls.append(url_temp)

    # Execução para o caso de encontrar erro
    except:

      # Comando para continuar a execução sem realizar ações
      pass
```

```python
# Exibe conteúdo da lista de urls dos estados e da lista de siglas
print(lista_de_urls)
print(lista_de_siglas)
```

```python
ind = 20
print(lista_de_urls[ind])
print(lista_de_siglas[ind])
```

```python
count = 1
# Cria loop para cara elemento das listas de siglas e de urls
for sigla, url_temp in zip(lista_de_siglas,lista_de_urls):

  # Realiza requisição na url do estado
  resposta_temp = requests.get(url_temp)

  # Converte a string da resposta em objeto 'soup'
  soup_temp = bs4.BeautifulSoup(resposta_temp.text)

  # Realiza a busca por elementos da classe 'even' no conteúdo de 'soup'
  links_temp = soup_temp.find_all( class_ = 'even')


  print("-" * 20)

  # Exibe a sigla do estado atual do loop
  print("\nLINKS DO ESTADO {} \n".format(sigla))


  # Realiza loop para cada link encontrado na página do estado
  for link in links_temp:
    count += 1
    # Adquire o conteúdo do elemento 'href'
    conteudo_link = link.contents[0]['href']

    base_url_temp = 'http://premiacao.obmep.org.br/16aobmep/'

    # Adiciona conteúdo do link à url base 
    print(count, base_url_temp + conteudo_link)

  print()

```

```python
# Biblioteca para manipulação de dados (dataframe)
import pandas as pd

# Pegamos um dos links para servir de base para a estrutura de todos os demais
resp_temp = requests.get('http://premiacao.obmep.org.br/16aobmep/verRelatorioPremiadosGeral-TO.1.do.htm')

# Convertermos a string do conteúdo da resposta em objeto 'soup'
soup_temp = bs4.BeautifulSoup(resp_temp.text)


# Adquire dados da tabela para extração do cabeçalho
header_data = soup_temp.find_all(class_ = "list")[0].find_all("tr")

# Cria lista vazia para cabeçalho
cabecalho = []


# Adiciona cada elemento do cabeçalho à lista
for h in header_data[1]:
  cabecalho.append(h.get_text())


# Adquire dados da tabela para extração dos alunos, escola, ...
HTML_data = soup_temp.find_all("tbody")[0].find_all("tr")


# Cria lista vazia para adicionar os dados
dados = []


# Loop para cada elemento da tabela
for element in HTML_data:

  # Cria lista vazia para subelementos
  temp = []

  # Loop para os subelementos da tabela
  for subelemento in element:
    
    # Adiciona subelementos à lista temporária
    temp.append(subelemento.get_text())
  
  # Adiciona lista temporária aos dados
  dados.append(temp)

print(cabecalho)
print(dados)

```

```python
# Cria dataframe com cabeçalho e conteudo
dataFrame = pd.DataFrame(data = dados, columns = cabecalho)

# Exporta dados para csv
dataFrame.to_csv('teste.csv')
```

```python
# Biblioteca para requisição via GET/POST
import requests

# Biblioteca para conversão de strings em objetos HTML, manipulação e filtros destes elementos
import bs4

# Biblioteca para manipulação de dados (dataframe)
import pandas as pd


# Declara String para requisição
ulr_base = 'http://premiacao.obmep.org.br/16aobmep/mapa.htm'

# Requisição no formato GET para a URL
resposta = requests.get(ulr_base)


# Converte String com conteúdo HTML para objeto SOUP que possibilita filtrar os elementos e obter resultados de pesquisa
soup = bs4.BeautifulSoup(resposta.text)

# Cria um objeto denominado 'select' contendo apenas os elementos encontrados dentro de 'soup', conforme o parâmetro para busca
select = soup.find_all('select')


# Cria duas listas vazias
lista_de_urls = []
lista_de_siglas = []

# Cria loop para cada elemento dentro do objeto 'select'
print("Criando links por Estado")
print("-" * 20)
for s1 in select:

  # Adquire o conteúdo de cada elemento em 'select -> s1'
  for s2 in s1:
    # Comando para testar um bloco de comandos e direcionar para o caso de erro
    try:

      # Obtém o conteúdo (caso exista) de cada elemento e concatena à url para cada estado.
      sigla = s2.get_text()

      # remove quebra de linha da string
      sigla = sigla.replace("\n", "")

      # Verifica se o conteúdo da string tem mais que ZERO caracteres
      if len(sigla) > 0:
        # Adiciona sigla à lista de siglas
        lista_de_siglas.append(sigla)

        # Adiciona a sigla do estado à string 'url_temp'
        url_temp = 'http://premiacao.obmep.org.br/16aobmep/verMenuAlunosPremiados-{}.htm'.format(sigla)

        print(url_temp)

        # Adiciona url_temp à lista
        lista_de_urls.append(url_temp)

    # Execução para o caso de encontrar erro
    except:

      # Comando para continuar a execução sem realizar ações
      pass


print("Processo concluído!")
print("Iniciando Extração de links de cada Estado")
print("-"*20)

# Cria loop para cara elemento das listas de siglas e de urls
for sigla, url_temp in zip(lista_de_siglas,lista_de_urls):

  # Realiza requisição na url do estado
  resposta_temp = requests.get(url_temp)

  # Converte a string da resposta em objeto 'soup'
  soup_temp = bs4.BeautifulSoup(resposta_temp.text)

  # Realiza a busca por elementos da classe 'even' no conteúdo de 'soup'
  links_temp = soup_temp.find_all( class_ = 'even')


  print("-" * 20)

  # Exibe a sigla do estado atual do loop
  print("\nLINKS DO ESTADO {} \n".format(sigla))


  # Realiza loop para cada link encontrado na página do estado
  for link in links_temp:
    
    # Adquire o conteúdo do elemento 'href'
    conteudo_link = link.contents[0]['href']

    base_url_temp = 'http://premiacao.obmep.org.br/16aobmep/'

    # Adiciona conteúdo do link à url base 
    print(base_url_temp + conteudo_link)

  print()

print("-"*20)
print("Processo concluído!")
print("Iniciando Extração de cada link")

# Cria loop para cara elemento das listas de siglas e de urls
for sigla, url_temp in zip(lista_de_siglas,lista_de_urls):

  # Realiza requisição na url do estado
  resposta_temp = requests.get(url_temp)

  # Converte a string da resposta em objeto 'soup'
  soup_temp = bs4.BeautifulSoup(resposta_temp.text)

  # Realiza a busca por elementos da classe 'even' no conteúdo de 'soup'
  links_temp = soup_temp.find_all( class_ = 'even')


  print("-" * 20)

  # Exibe a sigla do estado atual do loop
  print("\nLINKS DO ESTADO {} \n".format(sigla))

  # Realiza loop para cada link encontrado na página do estado
  for link in links_temp:

    # Rotina de teste
    try:
      # Adquire o conteúdo do elemento 'href'
      conteudo_link = link.contents[0]['href']

      base_url_temp = 'http://premiacao.obmep.org.br/16aobmep/'

      # Adiciona conteúdo do link à url base 
      print(base_url_temp + conteudo_link)


      resp_temp = requests.get(base_url_temp + conteudo_link)


      # Convertermos a string do conteúdo da resposta em objeto 'soup'
      soup_temp = bs4.BeautifulSoup(resp_temp.text)


      # Adquire dados da tabela para extração do cabeçalho
      header_data = soup_temp.find_all(class_ = "list")[0].find_all("tr")

      # Cria lista vazia para cabeçalho
      cabecalho = []


      # Adiciona cada elemento do cabeçalho à lista
      if 'Nome' in str(header_data[0]) and 'UF' in str(header_data[0]):
        for h in header_data[0]:
          cabecalho.append(h.get_text())

      elif 'Nome' in str(header_data[1]) and 'UF' in str(header_data[0]):
        for h in header_data[1]:
          cabecalho.append(h.get_text())

      elif 'Tipo' in str(header_data[2]) and 'UF' in str(header_data[0]):
        for h in header_data[2]:
          cabecalho.append(h.get_text())

      # Adquire dados da tabela para extração dos alunos, escola, ...
      HTML_data = soup_temp.find_all("tbody")[0].find_all("tr")


      # Cria lista vazia para adicionar os dados
      data = []


      # Loop para cada elemento da tabela
      for element in HTML_data:

        # Cria lista vazia para subelementos
        temp = []

        # Loop para os subelementos da tabela
        for subelemento in element:

          # Adiciona subelementos à lista temporária
          temp.append(subelemento.get_text())
        
        # Adiciona lista temporária aos dados
        data.append(temp)
    # Em caso de erro, pula o ciclo no loop
    except:
      continue

    try:
      # Cria dataframe com cabeçalho e conteudo
      dataFrame = pd.DataFrame(data = data, columns = cabecalho)

      # Exporta dados para csv
      dataFrame.to_csv('bd/{}.csv'.format(conteudo_link.split('.')[0].replace("ver","")))

    # Em caso de erro continua a execução
    except:
      pass
```
